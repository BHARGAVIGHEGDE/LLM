```python
import requests
import re

class Website:
    def __init__(self, url):
        self.url = url  

    def fetch(self):
        response = requests.get(self.url)
        self.html = response.text

    def parse(self):
        # extract title
        title_match = re.search(r"<title>(.*?)</title>", self.html, re.I | re.S)
        self.title = title_match.group(1).strip() if title_match else "No title"

        # remove tags (basic cleanup)
        clean = re.sub(r"<script.*?>.*?</script>", "", self.html, flags=re.S)
        clean = re.sub(r"<style.*?>.*?</style>", "", clean, flags=re.S)
        clean = re.sub(r"<[^>]+>", "", clean)

        self.text = clean.strip()
```

```python
bh=Website("https://docs.astral.sh")
bh.fetch()
bh.parse()
print(bh.title)
print(bh.text)
```

```python
system_prompt= "You are an assistant that analyzes the contents of a website \
and provides a short summary, ignoring text that might be navigation related. \
Respond in markdown."
```

```python
def user_prompt(bh):
    user_prompt="You are looking at a website titled {website.title} The contents of this website is as follows; \
please provide a short summary of this website in markdown. \
If it includes news or announcements, then summarize these too.\n\n"
    
    return user_prompt
```

```python
messages=[{"role":"system","content":"you are nerd"},
          {"role":"user","content":"are you nerd?"}]
```

```python
response=openai.chat.completions.create(model="llama3.2",messages=messages)
print(response.choices[0].message.content)
```

```python
def messages_for(website):
    return[
        {"role":"system","content":system_prompt},{
        "role":"user","content":user_prompt(website)
        }
    ]
```

```python
def summarize(url):
    website=Website(url)
    response=openai.chat.completions.create(
        model="llama3.2",messages=messages_for(website)
    )
    return response.choices[0].message.content
```

```python
summarize("https://docs.astral.sh")
```
